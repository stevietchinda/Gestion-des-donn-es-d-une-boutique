```python
#premiere vu des donnÃ©es ERP
# Aucune valeur nulles (NaN)
#Pas d'erreur lexical

data_erp =pd.read_csv('erp.csv')
data_erp.info()
data_erp.head()
```

```python
#VÃ©rification des valeurs nulles
# pas de valeurs nulles

print(data_erp.isnull().sum())
```

```python
# VÃ©rification des doublons
# Pas de doublons

data_erp.loc[data_erp['product_id'].duplicated(keep=False),:]
```

```python
# Recherche des erreurs de saisie
# Filtre les produits qui sont marquÃ©s comme Ã©tant en stock ('instock').
# Identifie les erreurs parmi ces produits, c'est-Ã -dire ceux qui ont une quantitÃ© en stock de 0.
# Le DataFrame errorinstock contient donc les produits qui sont incorrectement marquÃ©s comme Ã©tant en stock, mais qui ont en fait une quantitÃ© de stock de 0.

instock = data_erp[data_erp['stock_status']== 'instock']
errorinstock = instock[instock['stock_quantity'] ==0] 

errorinstock
```

```python
# Recherche des erreurs de saisie

outofstock = data_erp[data_erp['stock_status']== 'outofstock']
erroroutofstock = outofstock[outofstock['stock_quantity'] !=0] 

erroroutofstock
```

```python
#  Correction des stocks
data_erp['stock_status'] = np.where(data_erp['stock_quantity']==0, 'outofstock', 'instock')
```

```python
# VÃ©rification si la correction des erreurs Ã  fonctionner

instock = data_erp[data_erp['stock_status']== 'instock']
errorinstock = instock[instock['stock_quantity'] ==0] 

errorinstock
```

```python
#Nettoyage des donnÃ©es
#  Aucune valeur nulles (NaN) dans la colonne â€˜product_idâ€™
# Des valeurs nulles (NaN) dans la colonne â€˜id_webâ€™

data_liaison =pd.read_csv('liaison.csv')
data_liaison.info()
```

```python
# recherche de valeurs nulles
# 91 valeurs nulles dans id_web

print(data_liaison.isnull().sum())
```

```python
# Suppression des valeurs nulles 

data_liaison = data_liaison.dropna()
#print(data_liaison.isnull().sum())
```

```python
#Recherche des doublons
#Pas de doublons

data_liaison.loc[data_liaison['id_web'].duplicated(keep=False),:]
```

```python
#Premier merge entre les donnÃ©es de liaison et les donnÃ©es ERP

dataLE = pd.merge(data_liaison, data_erp, how='inner', on=['product_id'])
dataLE                  
```

```python
#DeuxiÃ¨me merge entre le premier merge et les donnÃ©es WEB

data = pd.merge(dataLE, data_web, how='inner', on=['id_web'])
data
```

```python
#Calcul du CA par produit et du CA total
#Ajout dâ€™une colonne â€˜CAâ€™ qui prÃ©sente le chiffre dâ€™affaire de chaque produits

data = data.assign(CA = data['price']*data['total_sales']) 
data[['product_id', 'id_web', 'price', 'total_sales', 'post_title', 'CA']]
```

```python
# Calcul du du CA total
print(data['CA'].sum())
```

```python
# Calcul de la moyenne
# Calcul de la mediane

print( "La moyenne est : ",data['price'].mean().round(2))
print("La mÃ©diane est : ", data['price'].median())

# Calculer Q1 (25e percentile)
Q1 = data['price'].quantile(0.25).round()

# Calculer Q3 (75e percentile)
Q3 = data['price'].quantile(0.75).round()

# Calculer IQR (Intervalle Interquartile)
IQR = (Q3 - Q1).round()

# Afficher les rÃ©sultats
print("Q1 :", Q1)
print("Q3 :", Q3)
print("IQR  :", IQR)

#minimun et maximum
print("min : ", data['price'].min())
print("max : ", data['price'].max())

```

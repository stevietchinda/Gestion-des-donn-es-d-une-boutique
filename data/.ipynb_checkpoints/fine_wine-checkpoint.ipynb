```python
# Renommer la colonne sku

data_web = data_web.rename(columns={'sku' : 'id_web' }) 
```

```python
data_web.head(3)
```

```python
data_web.info(1)
```

```python
# Crée un DataFrame booléen indiquant la présence de valeurs nulles.
# Calcule le nombre de valeurs nulles dans chaque colonne.
# Affiche ce nombre pour chaque colonne du DataFrame.
print(data_web.isnull().sum())
```

```python
#  Remplacement des 0 du dataFrame par NaN

import numpy as np  # Pour utiliser np.nan qui représente les valeurs manquantes.
data_web=data_web.replace(0,np.nan)
print(len(data_web.index), 'lignes avant la suppression des NaN') # Renvoie le nombre de lignes dans le DataFrame data_web.
data_web.head()
      
```

```python
#Suppression des lignes NaN

data_web =data_web.dropna(how= 'all') 
print(len(data_web.index), 'lignes après la suppression des NaN') # Renvoie le nombre de lignes dans le DataFrame data_web.
data_web.head()
```

```python
# Conversion de la list en DF

print(type(data_web)) # pour verifier le type de donnée
data_web_df = pd.DataFrame(data_web)  
print(type(data_web_df))
data_web_df.head()
#data_web_null = data_web[data_web['id_web'].isnull()]

#print(data_web_null)
```

```python
# Recherche des ‘id_web’ nulles (NaN)
#data_web_df = pd.DataFrame(data_web, columns=['id_web'])

data_web_null = data_web_df[data_web_df['id_web'].isnull()]
data_web_null.head()
```

```python
# Suppression des produits ayant pour ‘id_web’ nulles (NaN)

data_web = data_web.dropna(subset=['id_web'])
```

```python
print(len(data_web.index), 'lignes après la suppression des NaN') # Renvoie le nombre de lignes dans le DataFrame data_web.
data_web.head()
```

```python
#Selection des données importantes

data_web = data_web[['id_web', 'total_sales', 'post_date', 'post_title', 'post_excerpt', 'post_name', 'post_type']]
data_web
```

```python
#permet d'isoler et d'afficher toutes les lignes du DataFrame data_web où la colonne 'id_web' contient des valeurs dupliquées.

data_web.loc[data_web['id_web'].duplicated(keep=False),:]
```

```python
# Suppréssion des doublons

data_web  = data_web.drop_duplicates(subset=['id_web'])
data_web
```

```python
# Remplacer les valeur nulles (Nan) de la colonne ‘total_sales’ par des zéro

data_web = data_web.replace(np.nan, 0)
data_web.sort_values(by=['id_web'], ascending=False)
```
